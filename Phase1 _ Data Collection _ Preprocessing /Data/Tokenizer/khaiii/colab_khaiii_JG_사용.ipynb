```
import pandas as pd
df = pd.read_pickle('/content/drive/My Drive/chatbot_for_the_disabled/Phase1 : Data Collection & Preprocessing /Data/통합데이터/total_df_26810.pickle')
```

```
from khaiii import KhaiiiApi
import re

api = KhaiiiApi()
def khaill_tokenizer(x):
  token_list = []
  for word in api.analyze(x):
    morphs_str = " + ".join([(m.lex + '/' + m.tag) for m in word.morphs])
    token_list.append(morphs_str)
  return token_list


def space_preprocessing(text):
    # 개행문자 제거
    text = re.sub(r'[\n\r\t]', '', str(text))
    return text

def special_preprocessing(text):
    # 개행문자 제거
    text = re.sub('[^ ㄱ-ㅣ가-힣 0-9 A-Z a-z]+', '', str(text))
    return text

df["Question"] = df["Question"].apply(space_preprocessing).apply(special_preprocessing).apply(lambda x : x.strip())
df["Answer"] = df["Answer"].apply(space_preprocessing).apply(special_preprocessing).apply(lambda x : x.strip())
```

```
df.drop(index = 9392, inplace = True)
df.drop(index = df[df['Answer'].apply(lambda x : len(x)) == 0].index, inplace = True)
```

```
df["Q_token"] = df["Question"].apply(khaill_tokenizer)
df["A_token"] = df["Answer"].apply(khaill_tokenizer)
```

```
df['len_Q_token'] = df['Q_token'].apply(lambda x : len(x))
df['len_A_token'] = df['A_token'].apply(lambda x : len(x))
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
fig, axes = plt.subplots(nrows = 1, ncols=2)
fig.set_size_inches(20, 5)
sns.distplot(df['len_Q_token'], ax = axes[0])
sns.distplot(df['len_A_token'], ax = axes[1])
```

```
df = df.reset_index(drop = True)
```
